```python
#This script is an attempt to sort through pulse oximetry values and categorize them by "Healthy", "Hyperventilation", "Hypoxemia", and "Severe" conditions


#Load in resources
import pandas as pd
import numpy as np
import ast
import random
import time

#Load in pulse oximetry csv file
raw_data = pd.read_csv("../Pulse Oximetry Data/Cleaned_datasets/pulseox_filtered.csv")
```

```python
#Isolating the oxygen saturation data from the csv file
o2sat = raw_data["saturation"]
```

```python
#Saving the oxygen saturation values into a 2xlen(o2sat) array
d = []
for x in range(0, len(o2sat)):
    entry = o2sat[x]
    d.append(entry)
```

```python
#Turning each str entry in d into a list entry in new array
clean_data = []
for x in range(0, len(d)):
    entry = d[x]
    clean_entry = ast.literal_eval(entry)
    clean_data.append(clean_entry)
```

```python
#For each list entry, I want to take the mean of each data
#And calculate the standard deviation for each population
#Save both values in separate arrays, their ordinality should be the same

mean_data = []

for x in range(0, len(clean_data)):
    #Calculating Mean (mu)
    n = 0
    list_entry = clean_data[x]
    for i in range(0, len(list_entry)):
        n = n + list_entry[i]
    mu = n / len(list_entry)

    #save to arrays
    mean_data.append(mu)
```

```python
#Need to sort the mean o2sat values by range into arrays

#Any value above 97 is hyperventilation
#any value between 95 and 97 is normal
#any value between 87 and 94 is hypoxemia
#Any value between 59 and 86 is severe hypoxemia

#Hyperventilation
hvent = []
#Healthy
healthy = []
#Hypoxemia
hypox = []
#Severe Hypoxemia
severe = []

for x in range(0, len(mean_data)):
    entry = round(mean_data[x])
    if entry > 97:
        hvent.append(entry)
    elif 95 <= entry <= 97:
        healthy.append(entry)
    elif 87 <= entry < 95:
        hypox.append(entry)
    elif 59 <= entry < 87:
        severe.append(entry)
    else:
        1
```

```python
#Saving the datasets as a csv

healthy_data = pd.DataFrame(healthy, columns=["healthy_values"])
healthy_data.to_csv("Healthy_OxygenSaturation.csv", index=False)

hyperventilation_data = pd.DataFrame(hvent, columns = ["hypervent_values"])
hyperventilation_data.to_csv("Hyperventilation_OxygenSaturation.csv", index=False)

hypoxemia_data = pd.DataFrame(hypox, columns=["hypoxemia_values"])
hypoxemia_data.to_csv("Hypoxemia_OxygenSaturation.csv", index=False)

severe_data = pd.DataFrame(severe, columns=["severe_values"])
severe_data.to_csv("Severe_OxygenSaturation.csv", index=False)
```
