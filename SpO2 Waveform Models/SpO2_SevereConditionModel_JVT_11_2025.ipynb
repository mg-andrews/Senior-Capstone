```python
#Severe Hypoxemia Oxygen Saturation Model

#Load in modules
import pandas as pd
import numpy as np
import ast
import random
import time
```

```python
#Load in csv file
severe_data = pd.read_csv("../SpO2 Waveform Models/Sorted Data and Sorting Function/Severe_OxygenSaturation.csv")
```

```python
#Reading csv file
severe_values = severe_data["severe_values"]
```

```python
#Standard Deviation
n = 0
s = 0
for x in range(0, len(severe_values)):
    n = n + severe_values[x]
mu = n / len(severe_values)
for x in range(0, len(severe_values)):
    s = s + (severe_values[x] - mu)**2
sd = round((s / len(severe_values))**0.5)
```

```python
#Severe Hypoxemia Oxygen Saturation Output
num = random.randint(0, len(severe_values))
output = severe_values[num]
upper_output = output + random.randint(1,2)
lower_output = output - random.randint(1,2)

#This is to make the output on the screen fluctuate like a real pulse ox
to_screen_outputs = [lower_output, output, upper_output, output]
```

```python
#Display output
for x in range(0, len(to_screen_outputs)):
    print(to_screen_outputs[x])
    time.sleep(3.2)
```
