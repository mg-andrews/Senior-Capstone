```python
#Hypoxemia Oxygen Saturation Model

#Load in modules
import pandas as pd
import numpy as np
import ast
import random
import time
```

```python
#Load in csv file
hpox_data = pd.read_csv("../SpO2 Waveform Models/Sorted Data and Sorting Function/Hypoxemia_OxygenSaturation.csv")
```

```python
#Reading csv file
hpox_values = hpox_data["hypoxemia_values"]
```

```python
#Standard Deviation
n = 0
s = 0
for x in range(0, len(hpox_values)):
    n = n + hpox_values[x]
mu = n / len(hpox_values)
for x in range(0, len(hpox_values)):
    s = s + (hpox_values[x] - mu)**2
sd = round((s / len(hpox_values))**0.5)
```

```python
#Hypoxemia Oxygen Saturation Output
num = random.randint(0, len(hpox_values))
output = hpox_values[num]
upper_output = output + sd
lower_output = output - sd

#This is to make the output on the screen fluctuate like a real pulse ox
to_screen_outputs = [lower_output, output, upper_output, output]
```

```python
#Display output
for x in range(0, len(to_screen_outputs)):
    print(to_screen_outputs[x])
    time.sleep(3.2)
```
