```python
#This script is an attempt to create a function to calculate the pessimistic and realistic time from an input optimistic time
#  More stuff will be added here as the design schedule progresses
#  Hopefully i want to import the matlab code dmitri gave us for probabilities

import numpy as np
def find_times(optimistic_time, delay_times, delay_probs, d1, d2):
    #optimistic_time is an integer with units "days"
    
    #delay times is a list of integers with units "days"
    #ex. delay_times = [1,3,4,8,2]
    #delay probs is a list of fractions between 0 and 1 (unitless)
    #ex. delay_probs = [0.2, 0.6, 0.5, 0.4, 0.2]
    #Make sure the order of numbers in delay_times is the same for delay_probs
    # delay_times[0] has the probability of delay_probs[0]

    #d1 and d2 are two element lists where d1[0] = most likely delay time and d1[1] = most likely delay prob.
    opt = optimistic_time
    pessimistic_time = 0
    n = 0
    
    if len(delay_times) == len(delay_probs):
        for x in range(0, len(delay_times)):
            n = n + delay_times[x] * delay_probs[x]
    pessimistic_time = opt + n

    M = optimistic_time + d1[0] * d1[1] + d2[0] * d2[1]

    exp_time = (optimistic_time + pessimistic_time + 4 * M) / 6

    print("The pessimistic time is: "+str(pessimistic_time)+" days")
    print("The most likely time is: "+str(M)+" days")
    print("The expexted time is: "+str(exp_time)+" days")
```
